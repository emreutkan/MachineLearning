# Linear Regression with Python
Linear regression is a statistical method that models the relationship between two variables by fitting a linear equation to the observed data. It is used to predict the value of a dependent variable based on the value of an independent variable. The relationship between the two variables is called the dependent and independent variables. The dependent variable is the variable that we want to predict, while the independent variable is the variable that we use to make the prediction.


The linear regression model can be represented by the equation: $$y = mx + b$$
where: 
- y is the dependent variable
- x is the independent variable
- m is the slope of the line
- b is the y-intercept

# Problem Statement 
Suppose we have a dataset that contains the size of houses in meter square and their prices. We want to build a linear regression model that can predict the price of a house based on its size. The dataset contains two columns: 
- Square Meters: the size of the house in meter square
- Price: the price of the house in dollars

# Steps to Build the Model
1. Data Preparation: load the dataset and prepare the data for building the model
2. Data Visualization: visualize the data to understand the relationship between the independent and dependent variables
3. Data Splitting: split the data into training and testing sets
4. Data Reshaping: reshape the data to be compatible with the linear regression model
5. Model Building: build the linear regression model using the training data
6. Model Evaluation: evaluate the model performance using the testing data
7. Model Visualization: visualize the model predictions
8. Conclusion: conclude the findings and the model performance

# Tools and Libraries
- Python: an open-source programming language
- Numpy: a library used for numerical computations
- Matplotlib: a library used for data visualization
- Scikit-learn: a library used for machine learning models
- Jupyter Notebook: an open-source web application that allows you to create and share documents that contain live code, equations, visualizations, and narrative text

# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

# Data Preparation

```python
np.random.seed(42) #

m2 = np.random.randint(120, 840, size=50) # dependent variable
price = 150000 + 250 * m2 + np.random.randint(-30000, 30000, size=50) # independent variable
```

# Data Visualization
you can use matplotlib or seaborn to visualize the data

```python
plt.scatter(m2, price)
plt.xlabel('Square Meters')
plt.ylabel('Price')
plt.title('Price vs Square Meters')
plt.grid(True)
plt.show()
```

# Data Splitting
to avoid overfitting and underfitting issues we will use 80% of the data for training and 20% for testing
we have to split the data into training and testing sets to evaluate the model performance on unseen data points

sklearn provides a function called train_test_split that can be used to split the data into training and testing sets

```python
m2_train, m2_test, price_train, price_test = train_test_split(m2, price, test_size=0.2, random_state=42)
```

# Data Visualization after splitting

```python
plt.scatter(m2_train, price_train, label='Training Data')
plt.scatter(m2_test, price_test, label='Testing Data')
plt.xlabel('Square Meters')
plt.ylabel('Price')
plt.title('Price vs Square Meters')
plt.legend()
plt.grid(True)
plt.show()
```

# Data Reshaping
the data should be reshaped before building the model using the linear regression model 
the reshape function is used to change the shape of the data to be compatible with the model

the reshape function takes the number of rows and columns as input and returns a new array with the specified shape

```python
m2_train = m2_train.reshape(-1, 1)
```

# Model Building
so far we have prepared the data and split it into training and testing sets
now we can build the model using the training data
and to do that we will use the linear regression model from sklearn
this will help us to find the best fit line that represents the relationship between the independent and dependent variables
using the training data we can fit the model and predict the price of the house based on the size of the house

```python
model = LinearRegression()
model.fit(m2_train, price_train)
```

# Model Evaluation
after building the model we can evaluate its performance using the testing data
we can use the predict function to predict the price of the house based on the size of the house

```python
price_pred = model.predict(m2_test.reshape(-1, 1))
```

# Model Visualization

```python
plt.scatter(m2_test, price_test, label='Actual Price')
plt.plot(m2_test, price_pred, color='red', label='Predicted Price')
plt.xlabel('Square Meters')
plt.ylabel('Price')
plt.title('Price vs Square Meters')
plt.legend()
plt.grid(True)
plt.show()
```

# Model Evaluation
from the graph alone we can see that the model is doing a good job in predicting the price of the house based on the size of the house
but we can also use some metrics to evaluate the model performance

1. Mean Absolute Error (MAE): the average of the absolute differences between the predicted and actual values
2. Mean Squared Error (MSE): the average of the squared differences between the predicted and actual values
3. R-squared (R2): the proportion of the variance in the dependent variable that is predictable from the independent variable

```python
mae = mean_absolute_error(price_test, price_pred)
mse = mean_squared_error(price_test, price_pred)
r2 = r2_score(price_test, price_pred)
print(f'Mean Absolute Error:     {int(mae)}')
print(f'Mean Squared Error:      {int(mse)}')
print(f'R-squared Score:         {r2}')
```

# Conclusion
the model is doing a good job in predicting the price of the house based on the size of the house
the R-squared score is close to 1 which means that the model can explain most of the variance in the dependent variable
the Mean Absolute Error and Mean Squared Error are also low which indicates that the model is making accurate predictions
